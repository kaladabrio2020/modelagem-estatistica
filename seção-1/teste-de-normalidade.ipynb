```r
library(data.table)
```

```r
test_ = function(x, nivel_significancia=0.05){
    if (x > nivel_significancia){
        paste0(paste0('Aceita H_0, para um nivel de significancia ', 
        toString.default(nivel_significancia)) , 
        ' os dados seguem uma normal' )
    }else{
        paste0(paste0('Rejeita H_0, para um nivel de significancia ',
        toString.default(nivel_significancia)),
        ' os dados n√£o seguem uma normal' )

    }
}
```
