```r
set.seed(10)
```

```r
x = rnorm(20, sd = 0.4, mean = 4)
y = x * x * .05 + 10
plot(x ,y )
```

```r
model = lm(y~x)
```

```r
model
```

```r
par(bg = 'white')
options(
    repr.plot.width = 15,    # Largura em polegadas
    repr.plot.height = 8,   # Altura em polegadas
    repr.plot.res = 300   # Resolução em dpi
)

plot(x, y, pch = 19)
lines(x, unname(predict(model)),col = 'red')
```
